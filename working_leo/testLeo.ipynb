# Considerazioni per lavorare su Regioni

Il problema della nostra analisi è che non siamo in possesso di tutti i comuni all'interno di una singola regione; d'altro canto, però, la protezione civile non rilascia i dati per i decessi all'interno delle province. Noi siamo a conoscenza del fatto che i nostri dati provengono da comuni con un numero anomalo di decessi; di conseguenza, si può bene ipotizzare che possediamo i comuni nei quali la morte da coronavirus ha avuto un impatto significativo sulle morti. 

Si pensi di considerare la Lombardia (ma l'analisi può essere facilmente estesa a qualsiasi regione). In questa, noi possediamo circa il 50% per comuni totali: a priori un'analisi che prenda in considerazione la regione sarebbe errata. Però, come riportato dall'ISTAT, questo 50% dei comuni è quello in cui vi è stato un **incremento anomalo** di morti. Si chiami $M_{tot}$ le morti totali nel 2020 in Lombardia nel mese di Marzo (quello di interesse); esso sarà composto dalla somma dei morti per comune:
$$ M_{tot} = \sum_{i} m_i $$
dove $m_i$ è il numero di morti nel comune $i$-esimo. In presenza di pandemia di **covid**, si possono spezzare la variabili in due componenti:

- Morti "standard", ovvero cause di decesso non inficiabili (direttamente o indirettamente) alla presenza della pandemia. Da un'analisi statistica degli anni precedenti, ci si aspetta che queste morti non siano statisticamente differenti. Si chiama questa componente $n_i$.
- Morti dovuti al *covid-19*; questi sono dei morti **anomali** rispetto agli anni precedenti, presenti a causa della pandemia. Di conseguenza, confrontando questi morti con quelli degli anni precedenti (2015-19), ci si aspetta di trovare una significativa differenza statistica. Si chiama questa componente $c_i$.

Di conseguenza, si può riscrivere il numero totale di decessi in Lombardia come:
$$ 
M_{tot} = \sum_i n_i + \sum_i c_i 
$$
Come detto prima, però, noi possediamo solamente metà dei comuni; per semplicità, identifichiamo quelli che possediamo con pedice $k$ e quelli che non possediamo col pedice $j$. Sappiamo inoltre (e questa è una speranza risposta nell'ISTAT) che i dati posseduti da noi $k$ sono gli unici ad avere decessi anomali; ovvero sono gli unici ad avere una componente dovuta al covid. Di consguenza, posso scrivere:
$$
M_{tot} = \sum_j n_j + \sum_k n_k + \sum_k c_k 
$$
L'obiettivo della nostra analisi è quello di stimare il valore di $\sum_k c_k $. Per fare questo, si devono dunque costruire dei modello statistici che riescano a stimare le componenti "standard", ovvero $\sum_j n_j + \sum_k n_k$.

Si dimostra di seguito come questo possa essere fatto anche senza la conoscenza dei comuni $j$ (come è il nostro caso). 
1. **Conoscenza di $j$**. In questo caso, tramite dei modelli statistici basati sugli anni precedenti, si ottiene $\hat X = \hat{\left(\sum_j n_j + \sum_k n_k\right)}$ stimatore per le morti standard nei primi 3 mesi dell'anno. Per poter conoscere dunque una stima dei morti di covid-19, si deve sottrarre al numero totale di morti $M_{tot}$, che è conosciuto, questo stimatore, per ottenere la stima $\hat C = \hat{\sum_k c_k }$:
$$
    \hat C = M_{tot} - \hat X
$$
2. **Non conoscenza di $j$**. In questo caso, noi possediamo solamente una somma "mancante", ovvero:
$$ M^\prime_{quasi-tot} =  \sum_k n_k + \sum_k c_k $$
    Si procede però in modo analogo: si stima, con modelli statistici, la componente "standard" (tramite stimatore $\hat X^\prime = \hat{\sum_k n_k}$, da cui trovo poi la stima per la componente di covid:
$$ 
     \hat C = M^\prime_{quasi-tot} - \hat X^\prime
$$
    Di conseguenza, sebbene io abbia mancanza di informazione per metà dei comuni, sono comunque in grado di stimare i decessi in eccesso.

Ovviamente, questa analisi si poggia sull'ipotesi che effettivamente noi siamo a conoscenza di tutti e $k$ i comuni con morti anomale; così è riportato dall'ISTAT, ma errori umani sono sempre possibili. Un'analisi più accurata dovrebbe sicuramente seguire il procedimento 1.
